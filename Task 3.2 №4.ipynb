```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot()

a = -5
b = 5

n = 20

x_c = [(a+b)/2 + (b-a) * np.cos((2*k-1) * np.pi/(2*n))/2 for k in range(1, n+1)]
x_c1_5 = [x*(-1.5) for x in x_c]
y_c1_5 = np.sinc(x_c1_5)

ax.scatter(x_c, y_c1_5,
          marker = 'o', s = 20,
          color = 'g'
          )


def lagranz(x,y,t):
         z=0
         for j in range(len(y)):
             p1=1; p2=1
             for i in range(len(x)):
                 if i==j:
                     p1=p1*1; p2=p2*1   
                 else: 
                     p1=p1*(t-x[i])
                     p2=p2*(x[j]-x[i])
             z=z+y[j]*p1/p2
         return z
    
xnew=np.linspace(np.min(x_c),np.max(x_c),40)
ynew=[lagranz(x_c,y_c1_5,i) for i in xnew]

r = {'x' : [x for x in xnew], 'y' : [y for y in ynew]}
result = pd.DataFrame(data = r)
result.to_csv(r'C:\\Users\\nsavvin\Desktop\\result.csv')

df1 = pd.read_csv(r'C:\\Users\\nsavvin\Desktop\\result.csv' ,sep = ',')
x1=np.array(df1['x'], dtype=float)
y1=np.array(df1['y'], dtype=float)
 
ax.plot(x1, y1, color = 'r', linewidth = 1, label = 'После интерполяции')

ax.plot(x_c, y_c1_5, color = 'g', linewidth = 1, label = 'До интерполяции')

ax.scatter(x1, y1,
          marker = 'o', s = 20,
          color = 'r'
          )
ax.grid(which = 'major', color = 'gray', linestyle = ':')
ax.legend(fontsize = 14)
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
ax.tick_params(labelsize = 14)
```

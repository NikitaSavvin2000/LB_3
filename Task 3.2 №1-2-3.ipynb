```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
 
df = pd.read_csv(
    r'C:\\Users\\nsavvin\\Desktop\\Лабы\\LB_3\\zadanie__Dannye_Do_Interpolyatsii_-sinc_1_5x (1).csv',sep = ';'
)
print(df[:10])
print(df.columns)
 
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot()
ax.scatter(df['x'], df['y'],
          marker = 'o', s = 10,
          color = 'b', facecolors = 'none'
          )
ax.set_xlabel('x', fontsize = 14)
ax.set_ylabel('y', fontsize = 14)
ax.tick_params(labelsize = 14)
 
x=np.array(df['x'], dtype=float)
y=np.array(df['y'], dtype=float)
 
i = np.linspace(np.min(x),np.max(x), 100)
 
def lagranz(x,y,t):
         z=0
         for j in range(len(y)):
             p1=1; p2=1
             for i in range(len(x)):
                 if i==j:
                     p1=p1*1; p2=p2*1   
                 else: 
                     p1=p1*(t-x[i])
                     p2=p2*(x[j]-x[i])
             z=z+y[j]*p1/p2
         return z
    
xnew=np.linspace(np.min(x),np.max(x),40)
ynew=[lagranz(x,y,i) for i in xnew]
 
r = {'x' : [x for x in xnew], 'y' : [y for y in ynew]}
result = pd.DataFrame(data = r)
result.to_csv(r'C:\\Users\\nsavvin\\Desktop\\Лабы\\LB_3\\result_x_y.csv')
 
ax.scatter(xnew, ynew,
          marker = 'o', s = 10,
          color = 'black'
          )
ax.plot(x, y, color = 'g', label = 'До интерполяции')
 
df1 = pd.read_csv(r'C:\\Users\\nsavvin\\Desktop\\Лабы\\LB_3\\result_x_y.csv' ,sep = ',')
x1=np.array(df1['x'], dtype=float)
y1=np.array(df1['y'], dtype=float)
 
ax.plot(x1, y1, color = 'r', linestyle = '--', linewidth = 1, label = 'После интерполяции')

ax.legend(fontsize = 14)
ax.grid(which = 'major', color = 'gray', linestyle = ':')
 
plt.show()
```

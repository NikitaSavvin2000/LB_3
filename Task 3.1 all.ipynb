```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot()
 
d = pd.read_csv(r'C:\\Users\\nsavvin\Desktop\\Лабы\\LB_3\\approximatsia_Task_data_EURUSD.csv', delimiter = ';')
 
d['Уровень номера ряда'] = [i for i in range(1, len(d['<TIME>'])+1)]
x = np.array(d['Уровень номера ряда']).reshape((-1, 1))
y = np.array(d['<CLOSE>'])
y1 = np.array(d['<CLOSE>'][:15])
x1 = np.array(d['Уровень номера ряда'][:15]).reshape((-1, 1))
 
model = LinearRegression()
 
 
x_new = np.append(x1, np.array([i for i in range(len(x)+1, len(x)+11)])).reshape((-1, 1))
 
one_reg = PolynomialFeatures(degree = 1)
x_one = one_reg.fit_transform(x1)
model_one = model.fit(x_one, y1)
one_R2 = model.score(x_one, y1)
x_one_new = one_reg.fit_transform(x_new)
 
y_one_new = model_one.predict(x_one_new)
 
ax.plot(x_new, y_one_new,
        color = 'g',
        linestyle = '-.',
        linewidth = 1,
        label = ('n=1, R^2' + str(round(one_R2, 3)))
       )
 
two_reg = PolynomialFeatures(degree = 2)
x_two = two_reg.fit_transform(x1)
model_two = model.fit(x_two, y1)
two_R2 = model.score(x_two, y1)
x_two_new = two_reg.fit_transform(x_new)
 
y_two_new = model_two.predict(x_two_new)
 
ax.plot(x_new, y_two_new,
        color = 'y',
        linestyle = '-.',
        linewidth = 1,
        label = ('n=2, R^2' + str(round(two_R2, 3)))
       )
 
three_reg = PolynomialFeatures(degree = 3)
x_three = three_reg.fit_transform(x1)
model_three = model.fit(x_three, y1)
three_R2 = model.score(x_three, y1)
x_three_new = three_reg.fit_transform(x_new)
 
y_three_new = model_three.predict(x_three_new)
 
ax.plot(x_new, y_three_new,
        color = 'r',
        linestyle = '-.',
        linewidth = 1,
        label = ('n=3, R^2' + str(round(three_R2, 3)))
       )
 
ax.plot(d['Уровень номера ряда'][:15], d['<CLOSE>'][:15],
        color = 'b',
        linewidth = 3,
        label = 'close'
       )
ax.plot(d['Уровень номера ряда'][14:], d['<CLOSE>'][14:],
        color = 'g',
        linewidth = 3,
        label = 'close'
       )
 
 
 
ax.set_xlabel('Day', size = 14)
ax.set_xlabel('EUR/USD', size = 14)
ax.tick_params(labelsize = 14)
ax.set_ylim(1.15, 1.23)
ax.legend(fontsize = 14)
ax.grid(which = 'major', color = 'gray', linestyle = ':')
plt.show()
```
